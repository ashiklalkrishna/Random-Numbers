A. Generate uniform random numbers

```python
import numpy as np
from matplotlib import pyplot as plt

N=1000000
bins=100
print ("\nN=", N, "\n")
rnum = np.random.uniform (0,10,N)

plt.hist(rnum, density = 'True', color = 'teal', bins=bins)
plt.title ("Unifrom random number generation")
plt.xlabel("Random number sequence")
plt.ylabel ("Frequency")
plt.show()
```

B. Generate Gaussian random numbers

```python
import numpy as np
from matplotlib import pyplot as plt

mu=0                                #mean
sigma=1                             #standard deviation
N=1000000
bins=100
print ("\nN=", N, "\n")
rgaus = np.random.normal(mu,sigma,N)

plt.hist(rgaus, density = 'True', color = 'darkslateblue', bins=bins)
plt.title ("Gaussian random number generation")
plt.xlabel("Random number sequence")
plt.ylabel ("Frequency")
plt.show()
```

C. Generating Generating random numbers that mimic a particle's position probability

```python
import numpy as np
from matplotlib import pyplot as plt
fig=plt.figure(figsize=(10,5))

L=1.0
N=100000

x=np.random.uniform(0,L,N)
y=np.random.uniform(0,1,N)
z=[]
p=np.empty(N)

for i in range (N):
  p[i]=(np.sin((np.pi*x[i])/L))**2
  if (y[i]<p[i]):
    z.append(x[i])

w=np.linspace(0,1,N)
W=(2/L)*(np.sin((np.pi*w)/L))**2
plt.plot (w,W, 'r-')
plt.hist(z, density = 'True', color = 'lime', bins=100)
plt.title ("Position porbability distribution of a particle in 1D box")
plt.xlabel("Random number sequence")
plt.ylabel ("Probability")
plt.legend (['P(x)', 'distribution of random numbers'])
plt.show()
```

D. Generating mock/fake data

```python
import numpy as np
from matplotlib import pyplot as plt
fig=plt.figure(figsize=(10,5))

h = 100
u = 50
g = -9.8
mu = 0

t=np.linspace (0,10,20)
H0=h+(u*t)+(0.5*g*t**2)
print('Mock/fake data:\n',H0)

H=[]
err=[]

for i in range (0, len(t)):
  sig=0.05*H0[i]
  e=np.random.normal(mu, sig, 1)
  dH=H0[i]+e[0]
  H.append(dH)
  err.append(e)
error=abs(H-H0)

plt.scatter(t,H0, color='black')
plt.errorbar (t, H0, yerr=error, fmt='.', color='black')
plt.title ("Generating mock/fake data")
plt.xlabel("Time (s)")
plt.ylabel ("Height (m)")
plt.legend (['Data', 'Error'])
plt.show()
```
